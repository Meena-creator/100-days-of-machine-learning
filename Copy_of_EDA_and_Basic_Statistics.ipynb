### üìÄ Basic Statistics

* **Count** ‚Äì How many data points?
* **Mean** ‚Äì Average value (e.g., average age of users)
* **Median** ‚Äì Middle value when sorted (better for skewed data)
* **Standard Deviation** ‚Äì Spread of values from the mean
* **Min/Max** ‚Äì Range of values (spot outliers)

üìå Use `df.describe()` to get all of this at once.

---

### üìå Basic Statistics Example

```python
data = [1, 2, 2, 3, 14]
mean = sum(data)/len(data)  # 4.4
```

* **Mean** = 4.4 seconds (pulled up by the outlier 14)
* **Median** = 2 seconds (better "typical" value)
* **Standard Deviation** = Larger due to outlier

üí° **Use median for skewed data** like income, web traffic, etc.

---

### üßÆ Understanding Distributions

* Is the data **normal** (bell-shaped)?
* Is it **skewed** (left or right)?
* Are there **multiple peaks** (bi-modal)?

Use visual tools:

* **Histograms**
* **Box plots**

> **Histogram** groups values into bins and shows how frequent each bin is.
> E.g., most website sessions are under 200 seconds, but there‚Äôs a long tail of longer ones ‚Üí right-skewed.

* Built on top of Matplotlib
* Easier to make **beautiful** plots
* Integrates well with Pandas

Examples:

* Correlation heatmaps
* Box plots

> Seaborn provides visually appealing defaults and statistical overlays (like confidence intervals).
